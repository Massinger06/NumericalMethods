```python
#Primera derivada: método 1
#Análisis de error:  No conozco la derivada analítica

def f(x):
  return math.sin(x)

x = 0.5
n = 50
h = 1

epsilon = 10**(-11)

y_ini = 0.0

for i in range (0,n):
  h = 0.25*h
  Dy = f(x+h)-f(x)
  y = Dy/h #derivada
  error = abs((y-y_ini)/y)
  if error > epsilon:
    print('{0:<5}{1:<25.15e}{2:<25.15e}{3:<25.15e}'.format(i,h,y,error))
    y_ini = y
  else:
    print('{0:<5}{1:<25.15e}{2:<25.15e}{3:<25.15e}'.format(i,h,y,error))
    break
```

```python
#Primera derivada: método 2
#Análisis de error:  conozco la derivada analítica

def f(x):
  #Función a derivar
  #return -106.8*np.arctan(np.exp(-2/x))*np.sin(np.sqrt(4+x**2))
  return math.sin(x)

def g(x):
  #Derivada de la función
  #return -106.8*(x*np.arctan(np.exp(-2/x))*np.cos(np.sqrt(4+x**2))/np.sqrt(4+x**2)
  #              +np.sin(np.sqrt(4+x**2))/(np.cosh(2/x)*x**2))
  return math.cos(x)

x = 0.5
n = 50
h = 1

old_error = 1000000

for i in range (0,n):
  h = 0.25*h
  Dy = 0.5*(f(x+h)-f(x-h))
  y = Dy/h
  new_error = abs(g(x)-y)
  if new_error > old_error:
    break
  else:
    print('{0:<5}{1:<25.15e}{2:<25.15e}{3:<25.15e}'.format(i,h,y,new_error))
    old_error = new_error
```

```python
#Segunda derivada de una función
#Análisis de error: conozco la segunda derivada de la función

def f(x):
  #función a derivar
  return math.sqrt(x)

def g(x):
  #segunda derivada de f(x)
  return -0.25*x**(-1.5)

x = 0.5
n = 30
h = 1

old_error = 1000000

for i in range (0,n):
  h = 0.25*h
  Dy = f(x+h)-2*f(x)+f(x-h)
  y = Dy/(h*h)
  new_error = abs(g(x)-y)
  if new_error > old_error:
    break
  else:
    print('{0:<5}{1:<25.15e}{2:<25.15e}{3:<25.15e}'.format(i,h,y,new_error))
    old_error = new_error
```
